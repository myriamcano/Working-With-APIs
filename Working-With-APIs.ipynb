```python
import random
from datetime import datetime

def get_random_fact(data_frame):
    all_facts = data_frame['facts']
    
    if not all_facts.empty:
        return random.choice(all_facts)
    else:
        return 'No facts available.'

# Get and print a random fact
random_fact = get_random_fact(dogfacts)
print(f'Dog Fact of the day: {random_fact}')
```
